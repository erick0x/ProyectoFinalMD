## Indicador de seguridad

```python
import pandas as pd

df = pd.read_csv("C:\\Users\\GoodM\\OneDrive\\Escritorio\\ErickPython\\Rest-Mex_con_analisis.csv")

df.head(10)
```

### Filtrado

```python
# Buscar palabras clave en Review o Título y crear un DataFrame conjunto
palabras_clave = ['seguridad','robo','inseguro','amenaza','peligro','robado','asalto','violento','policía','seguro','ratero','alerta']
col_texto = None
for col in df.columns:
    if 'review' in col.lower() or 'comentario' in col.lower() or 'text' in col.lower():
        col_texto = col
        break
col_titulo = None
for col in df.columns:
    if 'title' in col.lower() or 'título' in col.lower():
        col_titulo = col
        break
if col_texto is None and col_titulo is None:
    print('No se encontró columna de texto ni de título')
else:
    filtro_texto = df[col_texto].str.contains('|'.join(palabras_clave), case=False, na=False) if col_texto else pd.Series([False]*len(df))
    filtro_titulo = df[col_titulo].str.contains('|'.join(palabras_clave), case=False, na=False) if col_titulo else pd.Series([False]*len(df))
    filtro_seguridad = filtro_texto | filtro_titulo
    df_seguridad = df[filtro_seguridad]
    print(f"Total de registros con palabras clave en título o review: {len(df_seguridad)}")
    display(df_seguridad.head(20))
```

### Media de sentimiento por cada pueblo

```python
# Calcular media de sentimiento_5 por pueblo (Town)
media_satisfaccion_por_town = (
    df_seguridad
        .groupby("Town")["sentimiento_5"]
        .mean()
        .sort_values(ascending=False)
)

print("Media de sentimiento_5 por Town:")
display(media_satisfaccion_por_town)
```

### Ponderación de 1-5

```python
media = media_satisfaccion_por_town

# Transformación lineal de [-1,1] → [0,5]
ponderacion_seguridad = ((media + 1) * 5) / 2

ponderacion_seguridad = ponderacion_seguridad.sort_values(ascending=False)

print("Ponderación 1–5 por Town:")
display(ponderacion_seguridad)
```

## Indicador de percepción climatica

### Filtrado

```python
import re

# Lista de palabras clave de clima
palabras_clave = ['clima','lluvia','tormenta','sol','temperatura','calor','frío','nieve','huracán','viento','inundación','tormentoso','nevado']

# Detectar columna de review
col_texto = None
for col in df.columns:
    if 'review' in col.lower() or 'comentario' in col.lower() or 'text' in col.lower():
        col_texto = col
        break

# Detectar columna de título
col_titulo = None
for col in df.columns:
    if 'title' in col.lower() or 'título' in col.lower():
        col_titulo = col
        break

# Filtrar reseñas de clima
if col_texto is None and col_titulo is None:
    print('No se encontró columna de texto ni de título')
else:
    filtro_texto = df[col_texto].str.contains('|'.join(palabras_clave), case=False, na=False) if col_texto else pd.Series([False]*len(df))
    filtro_titulo = df[col_titulo].str.contains('|'.join(palabras_clave), case=False, na=False) if col_titulo else pd.Series([False]*len(df))
    filtro_clima = filtro_texto | filtro_titulo
    df_clima = df[filtro_clima].copy()

    # Función para extraer ventana de 5 palabras alrededor de las palabras clave
    def extraer_contexto(texto, palabras_clave, ventana=5):
        resultados = []
        if pd.isna(texto):
            return resultados
        palabras = texto.split()
        for i, palabra in enumerate(palabras):
            for key in palabras_clave:
                if re.fullmatch(key, palabra, re.IGNORECASE):
                    start = max(i - ventana, 0)
                    end = min(i + ventana + 1, len(palabras))
                    contexto = ' '.join(palabras[start:end])
                    resultados.append(contexto)
        return resultados

    # Crear columna de contextos solo en df_clima
    if col_texto:
        df_clima['contexto_clima'] = df_clima[col_texto].apply(lambda x: extraer_contexto(x, palabras_clave))
```

### Quedarme con solo aquellas reseñas que tienen un contexto de clima

```python
# Filtrar solo los registros que tienen al menos un contexto
df_clima = df_clima[df_clima['contexto_clima'].map(len) > 0].copy()

print(f"Total de registros con contexto de palabras clave de clima: {len(df_clima)}")
display(df_clima.head(20))
```

### Media de sentimiento por cada pueblo

```python
# Calcular media de sentimiento_5 por pueblo (Town)
media_satisfaccion_por_town = (
    df_clima
        .groupby("Town")["sentimiento_5"]
        .mean()
        .sort_values(ascending=False)
)

print("Media de sentimiento_5 por Town:")
display(media_satisfaccion_por_town)
```

### Ponderación de 1-5

```python
media = media_satisfaccion_por_town

# Transformación lineal de [-1,1] → [0,5]
ponderacion_clima = ((media + 1) * 5) / 2

ponderacion_clima = ponderacion_clima.sort_values(ascending=False)

print("Ponderación 1–5 por Town:")
display(ponderacion_clima)
```

```python
import pandas as pd

# Leer el CSV existente
df = pd.read_csv('indicadores.csv', index_col='Town')

# Agregar las nuevas series
df['Percepcion_Seguridad'] = ponderacion_seguridad.round(2)
df['Percepcion_Clima'] = ponderacion_clima.round(2)

# Guardar nuevamente el CSV
df.to_csv('indicadores_completo.csv', index_label='Town')

# Verificar primeras filas
display(df.head())

# Opcional: verificar nulos
print("Valores nulos por columna:")
print(df.isnull().sum())
```
